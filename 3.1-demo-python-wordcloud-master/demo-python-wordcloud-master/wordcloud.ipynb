```python
filename = "Friends.txt"
with open(filename,encoding="utf8") as f:
    mytext = f.read()
```

```python
mytext
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud().generate(mytext)
```

```python
%pylab inline
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
