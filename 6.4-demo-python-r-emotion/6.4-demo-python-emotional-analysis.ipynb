```python
import pandas as pd  
import re
```

```python
with open('s03e09.txt',encoding='UTF-8') as f:  
    data = f.read()
```

```python
print(data)
```

```python
data = data.split('Opening Credits]')[1]
```

```python
print(data)
```

```python
data = data.split('[End Credits')[0]
```

```python
print(data)
```

```python
regex = r"^$\n"  
subst = ""  
data = re.sub(regex, subst, data, 0, re.MULTILINE) 
regex = r"^-+$\n"  
subst = ""  
data = re.sub(regex, subst, data, 0, re.MULTILINE) 
lines = data.split('\n') 
```

```python
myrows = []  
num = 1   
for line in lines:  
    myrows.append([num, line])  
    num = num + 1  
```

```python
myrows[:3]
```

```python
df = pd.DataFrame(myrows)
```

```python
df.head()
```

```python
df.columns = ['line', 'text']
```

```python
df.to_csv('data.csv', index=False)
```
